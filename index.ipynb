```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy
from sklearn.metrics import r2_score
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kecerdasan Buatan - Semester4/traffic.csv')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
x = []

for i in range(0,23):
  x.append(i)
```

# Data dari timur

```python
data_timur = df[['DateTime','Junction','Vehicles']].query('Junction == 1')
data_timur['DateTime'] = pd.to_datetime(data_timur['DateTime'])

data_timur_daily = []

for i in range(0, 23):
  data_timur_daily.append(data_timur[data_timur['DateTime'].dt.hour == i])
```

```python
data_timur_y = []

for i in range(0, 23):
  total = 0

  for num_of_vehicles in data_timur_daily[i]['Vehicles']:
    total += num_of_vehicles

  data_timur_y.append(total/len(data_timur_daily[i]['Vehicles']))

print(data_timur_y)
```

```python
plt.scatter(x, data_timur_y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, data_timur_y)

def predict(x):
  return slope * x + intercept

model_timur = numpy.poly1d(numpy.polyfit(x, data_timur_y, 3))

line_timur = numpy.linspace(0, 23, 100)

plt.scatter(x, data_timur_y)
plt.plot(line_timur, model_timur(line_timur))
plt.show()
```

```python
print(r2_score(data_timur_y, model_timur(x)))
```
